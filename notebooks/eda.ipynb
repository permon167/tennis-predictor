Cargar el archivo filtered.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('../data/processed/filtered.csv')
```

Configuración de visualización


```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

Estadísticas descriptivas

```python

print("\nEstadísticas descriptivas:")
print(data.describe())
```

Distribución de la variable objetivo (Winner)


```python
print("\nDistribución de la variable objetivo (Winner):")
winner_counts = data["Winner"].value_counts()
sns.barplot(x=winner_counts.index, y=winner_counts.values, palette="viridis")
plt.title("Distribución de la variable objetivo (Winner)")
plt.xlabel("Ganador")
plt.ylabel("Cantidad de partidos")
plt.show()
```

Matriz de correlación


```python
print("\nMatriz de correlación:")
# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación")
plt.show()
```

```python
# Distribución de la diferencia de ranking (rank_diff)
print("\nDistribución de la diferencia de ranking (rank_diff):")
sns.histplot(data["rank_diff"], kde=True, bins=30, color="blue")
plt.title("Distribución de la diferencia de ranking (rank_diff)")
plt.xlabel("Diferencia de ranking")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Análisis de la superficie del partido
print("\nDistribución de partidos por superficie:")
surface_counts = data["Surface"].value_counts()
sns.barplot(x=surface_counts.index, y=surface_counts.values, palette="muted")
plt.title("Distribución de partidos por superficie")
plt.xlabel("Superficie")
plt.ylabel("Cantidad de partidos")
plt.show()
```

```python
# Relación entre rendimiento reciente y victorias
print("\nRelación entre rendimiento reciente y victorias:")
sns.boxplot(x=(data["Winner"] == data["Player_1"]).astype(int), y=data["recent_form_p1"], palette="Set2")
plt.title("Rendimiento reciente del Jugador 1 vs. Resultado")
plt.xlabel("Jugador 1 Gana (1 = Sí, 0 = No)")
plt.ylabel("Rendimiento reciente (recent_form_p1)")
plt.show()

sns.boxplot(x=(data["Winner"] == data["Player_2"]).astype(int), y=data["recent_form_p2"], palette="Set3")
plt.title("Rendimiento reciente del Jugador 2 vs. Resultado")
plt.xlabel("Jugador 2 Gana (1 = Sí, 0 = No)")
plt.ylabel("Rendimiento reciente (recent_form_p2)")
plt.show()
```

```python

# Relación entre diferencia de ranking y victorias
print("\nRelación entre diferencia de ranking y victorias:")
sns.boxplot(x=(data["Winner"] == data["Player_1"]).astype(int), y=data["rank_diff"], palette="coolwarm")
plt.title("Diferencia de ranking vs. Resultado")
plt.xlabel("Jugador 1 Gana (1 = Sí, 0 = No)")
plt.ylabel("Diferencia de ranking (rank_diff)")
plt.show()
```
